```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
#from pandas import DataFrame 
import time

from sklearn.model_selection import train_test_split#导入数据分割包
```

```python
#设置字符集防止中文乱码
mpl.rcParams['font.sans-serif']=[u'simHei']
mpl.rcParams['axes.unicode_minus']=False
```

```python
#加载数据
df = pd.read_csv('data/household_power_consumption_1000.txt',sep = ';',low_memory=False)#low_memory=False没有混合类型时设置为False会提到读取效率
df.head(3)
```

```python
X = df.iloc[:,2:4]#取表示功率 的 Global_active_power	Global_reactive_power两列
Y = df.iloc[:,5]#取电流列
```

```python
#将数据分割为两部分 训练集和测试集
X_train,X_test = train_test_split(X,test_size = 0.2,random_state=0)#功能：数据集按比例切分为训练集和测试集

Y_train,Y_test = train_test_split(Y,test_size = 0.2,random_state=0)#random_state=0使随机划分的结果确定下来
type(X_train)
#Y_train.head()
```

```python
X_mat = np.mat(X_train)
Y_mat = np.mat(Y_train).reshape(-1,1)
```

```python
type(Y_mat)
```

```python
theta = (X_mat.T * X_mat).I * X_mat.T * Y_mat#求θ

theta
```

```python
#进行测试
Y_yuce = np.mat(X_test) * theta
len(Y_yuce)
```

```python
t=np.arange(len(X_test))
plt.figure(facecolor='r')
plt.plot(t, Y_test, 'r-', linewidth=2, label=u'真实值')
plt.plot(t, Y_yuce, 'b-', linewidth=2, label=u'预测值')
plt.legend(loc = 'upper right')#图例位置
plt.title(u"线性回归预测功率与电流之间的关系", fontsize=20)
plt.grid(b = True)
plt.show()
```
